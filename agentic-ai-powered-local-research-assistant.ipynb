```python
import fitz  # PyMuPDF
from transformers import pipeline
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain.text_splitter import RecursiveCharacterTextSplitter
import os
```

```python
import pymupdf
def load_pdfs(pdf_folder):
    """
    Load all PDFs in a folder and return a list of text chunks.
    """
    pdf_files = [os.path.join(pdf_folder, f) for f in os.listdir(pdf_folder) if f.endswith(".pdf")]
    print(f"Found {len(pdf_files)} PDFs:", pdf_files)
    
    all_texts = []
    for file_path in pdf_files:
        doc = pymupdf.open(file_path)
        for page in doc:
            text = page.get_text("text")
            if text.strip():
                all_texts.append(text)
    return all_texts

# Example: folder containing PDFs
pdf_folder = "/kaggle/input/skin-cancer-folder"
raw_texts = load_pdfs(pdf_folder)
```

```python
# Chunk text for vector store
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
docs = text_splitter.split_text(" ".join(raw_texts))
```

```python
all_chunks = []
for text in raw_texts:
    chunks = text_splitter.split_text(text)
    all_chunks.extend(chunks)
docs = all_chunks
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db = FAISS.from_texts(docs, embeddings)
retriever = db.as_retriever(search_kwargs={"k": 4})
```

```python
from transformers import pipeline

# Use a generative QA model
qa_generator = pipeline("text2text-generation", model="google/flan-t5-large", device=0)  # GPU
summarizer = pipeline("summarization", model="facebook/bart-large-cnn", device=0)
```

```python

def agentic_qa(query, mode="answer"):
    retrieved_docs = retriever.invoke(query)
    context = " ".join([doc.page_content for doc in retrieved_docs])

    if mode == "answer":
        prompt = f"Answer the following question based on the context below in a complete sentence.\n\nContext: {context}\n\nQuestion: {query}"
        result = qa_generator(
            prompt,
            max_new_tokens=150,
            do_sample=False
        )[0]["generated_text"]
        return result

    elif mode == "summary":
        summary = summarizer(
            context,
            max_length=120,
            min_length=40,
            do_sample=False
        )[0]["summary_text"]
        return summary

    elif mode == "keywords":
        return [word for word in query.split() if word.lower() not in ["what", "is", "the", "of", "in"]]

    else:
        return "⚠️ Mode not supported. Use: answer | summary | keywords"
```

```python
# ------------------ Interactive QA Loop ------------------
while True:
    query = input("\nAsk a question (or type 'exit' to quit): ")
    if query.lower() == "exit":
        break

    answer = agentic_qa(query, mode="answer")
    summary = agentic_qa(query, mode="summary")
    keywords = agentic_qa(query, mode="keywords")

    print("------------------------------------------------------------")
    print("Question:", query)
    print("Answer:", answer)
    print("Summary:", summary)
    print("Keywords:", keywords)
    print("------------------------------------------------------------")
```
